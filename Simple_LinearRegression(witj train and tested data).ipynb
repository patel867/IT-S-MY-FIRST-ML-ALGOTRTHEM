# Simple Linear Regression with Train and Test Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# loading dataset into dataset object
dataset=pd.read_csv("F:\machine learning\Simple linear Regression model_CSV__.csv")
```

```python
#retrieve only 0th column and take it as x
x=dataset.iloc[:,-1].values
x
#retrieve only 1st column and take it as y
y=dataset.iloc[:,1].values
y
```

*The above line retrieves(fetch or obtain) the values from the last column  and second column of the 'dataset' DataFrame using the `iloc` method. It selects all rows and the last column (-1 refers to the last column in Python) and also selects (index 1). The `.values` attribute returns the selected data as a NumPy array and assigns it to the variable 'x' and 'y'.

*.iloc method in pandas allows you to retrieve data from a DataFrame based on integer positions. Itâ€™s like using coordinates to pinpoint specific rows and columns

```python
#data scatter plot to verify simple linear regression model can be used.scatter plot show dot as straight line
plt.scatter(x,y)
```

```python
#take 70% of data for training and 30% for testing ,random_state indicates the random seed used in selecting test rows
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

above line calls the `train_test_split` function to split the input features `x` and the target variable `y` into training and testing sets.

```python
#train the computer  with simple linear regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()  # created a class object
# reshape the input data to 2D array
x_train_reshaped = x_train.reshape(-1, 1)
x_test_reshaped = x_test.reshape(-1, 1)
# train the model by passing train data to fit() method
reg.fit(x_train_reshaped, y_train)
# test the model by passing the test data and obtain the prediction
y_pred = reg.predict(x_test_reshaped)
y_pred
```

The `reshape` method is used to change the shape of the array. In this case, the input data `x_train` and `x_test` is a 1D array, but the `LinearRegression` model expects a 2D array as input. Reshaping the input data using the `reshape` method converts it into a 2D array, which is the expected format for the input to the `LinearRegression` model.

```python
#fnding the r squared value by comparing test data
#(excepted data) and predicted data
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#predicting the salary of employee with experience 11 yrs
print(reg.predict([[11]]))
```

```python
#using matplotlib, draw scatter plot and line plot
#select the below all statement and then execute at once
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train.reshape(-1,1)),color='blue')
plt.title("Experience vs Salary")
plt.xlabel('yrs of exp')
plt.ylabel('salary')
plt.show()
```
